```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("/content/drive/MyDrive/Assignment1_Q2_Data.csv", usecols=['House ID','Local Price','Bathrooms','Land Area','Living area','# Garages','# Rooms','# Bedrooms','Age of home','Construction type','Architecture type','# Fire places','Price (1000$)'])
np_data = data.to_numpy()
np_data = np.transpose(np_data)

x = np_data[1:12] 
y = np_data[12]
x_train = []
y_train = []
x_test = []
y_test = []
#training data loop
for i in range(11):
  x_train_temp = x[i]
  x_train.append(x_train_temp[0:18])
#test data loop
for i in range(11):
  x_test_temp = x[i]
  x_test.append(x_test_temp[18:28])
  
y_train = y[0:18]
y_test = y[18:28]
x_train = np.transpose(x_train)
x_test = np.transpose(x_test)


## theta learning
theta = np.matmul(np.linalg.inv(np.matmul(np.transpose(x_train),x_train)) , np.matmul(np.transpose(x_train),y_train) )

y_predict = np.matmul(x_test,theta)
y_train_predict = np.matmul(x_train,theta)

print('Theta values are = ',theta)
print('\n')

### Average Least Square Error
MSE_test_data = 0
for i in range(10):
  MSE_test_data = MSE_test_data + (y_predict[i] - y_test[i])**2

MSE_test_data = MSE_test_data/10
print('Average prediction error = ',MSE_test_data)


MSE_train_data = 0
for i in range(18):
  MSE_train_data = MSE_train_data + (y_train_predict[i] - y_train[i])**2

MSE_train_data = MSE_train_data/18
print('Average training data prediction error = ',MSE_train_data)


######### Plotting

test_data_points = list(range(10))
train_data_points = list(range(18))

print('\n \n')
# Prediction plot
plt.figure(1)
plt.plot(test_data_points,y_predict,'--b',label ='Prediction')
plt.plot(test_data_points,y_test,'*r', label = 'Test data')
plt.title('Prediction on test data')
plt.legend(loc='upper right')
plt.ylabel('Price * $1000')
plt.xlabel('Data point')


# Training data prediction
plt.figure(2)
plt.plot(train_data_points,y_train_predict,'--b',label ='Prediction on train data')
plt.plot(train_data_points,y_train,'*r', label = 'Training data')
plt.title('Prediction on Training Data')
plt.legend(loc='upper right')
plt.ylabel('Price * $1000')
plt.xlabel('Data point')

plt.show
```
